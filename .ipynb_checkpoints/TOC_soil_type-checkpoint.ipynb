```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ruta_site = ("../../data/raw/openagrar_derivate_00028499/BZE_LW English Version/SITE.xlsx")
ruta_lab = ("../../data/raw/openagrar_derivate_00028499/BZE_LW English Version/LABORATORY_DATA.xlsx")

df_site = pd.read_excel(ruta_site)
df_lab = pd.read_excel(ruta_lab)
```

1. Escogemos las variables para el análisis:
   SITE: PointID, coordinates, land use, Main soil type, Specific soil subtype
   LABORATORY_DATA: TOC, TN, BD_bulk, Water content


```python
# Escogemos las variables de ambos datasets
variables_site = ['PointID', 'xcoord', 'ycoord', 'Land use', 'Main soil type', 'Specific soil subtype']
variables_lab = ['PointID','TOC','TIC', 'TC', 'TN', 'BD_bulk', 'Water content', 'pH_H2O']

df_site2 = df_site[variables_site]
df_lab2 = df_lab[variables_lab]

df_lab2.head(10)
```

Antes de comenzar, necesitamos procesar los datos de LABORATORY_DATA:
- Como en df_lab2 tenemos los datos según horizontes agruparemos los suelos segun su PointID, por lo que el valor del TOC será la media de sus horizontes hasta la profundidad de 1 metro.

```python
grouped_lab = df_lab2.groupby('PointID', as_index=False).mean(numeric_only=True)
grouped_lab
```

```python
# Ambas tienen 3104 filas. Ya podemos unir ambos dataframes con la columna en común PointID:

df = pd.merge(df_site2, grouped_lab, on='PointID', how='outer')
df
```

4. EDA:
      Histograma de TOC por tipo de suelo.
4.1. Hacer un dataframe en el que tengamos en una columna el tipo de suelo y en la otra TOC
4.2. Agrupar pòr tipo de suelo, dando como valor al atributo TOC lla media de todas sus mediciones por tipo de suelo.


```python
df_tipos_TOC = df[ ['Main soil type', 'TOC'] ]
df_tipos_TOC
```

```python
df_tipos_TOC_grouped = pd.DataFrame(df_tipos_TOC.groupby('Main soil type')['TOC'].mean())
df_tipos_TOC_grouped
```

```python
# Histogram relative frequences 
df_tipos_TOC_grouped.plot(kind='bar', figsize=(12,6), color = 'brown')
plt.title('TOC mean in each main soil type')
plt.xlabel('Main soil type')
plt.ylabel('TOC (gr/kg)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
top10_TOC = df_tipos_TOC_grouped['TOC'].nlargest(10)
top10_TOC
```

Conclusion

As we can see, these results agree to a greater or lesser extent with the descriptions found in the International Soil Classification System for Naming Soils.

In studies such as Tobiášová et al. (2013), it is indicated that Chernozems can be classified based on the parameter of lability of carbon (LC), where the most labile carbon was observed in the following order: forest ecosystem (0.209) > agro-ecosystem (0.178) > meadow ecosystem (0.119) > urban ecosystem (0.116). Therefore, in agricultural ecosystems, these soils are particularly relevant for carbon monitoring, as they present a higher potential for carbon loss.

It is also worth mentioning that the same study (Tobiášová et al., 2013) highlights the inverse relationship between soil sand content and total organic carbon (TOC), which appears to be consistent with the findings from the description of this dataset.

This descriptive analysis does not pay particular attention to the dates and seasons when the data were collected. This is an important aspect, as both TOC and total nitrogen (TN) are known to undergo significant seasonal fluctuations (Kravchenko et al., 2022).
